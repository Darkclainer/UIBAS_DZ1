## Задача 3.
Имеется 5 билетов стоимостью по 1 рублю, 3 билета по 3 рубля и 2 билета по 5 рублей. Наугад берутся два билета. Найти вероятность того, что они имеют одинаковую стоимость.

### Решение.
Предпологаем, что все билеты одинаковой стоимости неотличимы друг от друга. Соотвественно порядок кв котором они вытягиваются - не имеет значения.

Событие "Вытянули два билета одинаковой степени" можно разбить на три *несовместных* события:
1. $A$ - Вытянули два билета стоимостью 1 рубль.
2. $B$ - Вытянули два билета стоимостью 3 рубля.
3. $C$ - Вытянули два билета стоимостью 5 рублей.

$P(A)$ состоит из двух событий: $A_1$ и $A_2$, которые должны произойти одно за другим. Событие $A_1$ состоит в том, что вытянули билет стоимостью 1 рубль. $P(A_1) = \frac{5}{5 + 3 + 2} = \frac{1}{2}$. Событие $A_2$ состоит в том, что второй билет, который вытянули также стоит 1 рублью. $P(A_2) = \frac{4}{4 + 3 + 2} = \frac{4}{9}$ - мы выбираем уже из 4 билетов, а всего билетов осталось 9, так как один билет уже вытянут.

$P(A) = P(A_1) \cdot P(A_2) = \frac{1}{2} \cdot \frac{4}{9} = \frac{2}{9}$

Аналогично можно посчитать $P(B)$ и $P(C)$.

$P(B) = \frac{3}{10} \cdot \frac{2}{9} = \frac{1}{15}$

$P(C) = \frac{2}{10} \cdot \frac{1}{9} = \frac{1}{45}$

Вероятность события $P(A \cup B \cup C) = P(A) + P(B) + P(C) = \frac{2}{9} + \frac{1}{15} + \frac{1}{45} = \frac{14}{45}$ &mdash; так как события несовместны. Это и есть искомая вероятность.

## Задача 4.
Вася неверно решает 10% задач, Петя - 15%. Каждому дано по задаче. 
Какова вероятность, что хотябы одна из двух задач решена неверно?

### Решение.
В данной задаче возможно всего четыре возможного исхода:

1. $R_{00}$ &mdash; Вася и Петя решили свои задачи неверно
2. $R_{01}$ &mdash; Вася решил свою задачу неверно, а Петя решил свою верно
3. $R_{10}$ &mdash; Вася верно решил свою задачу, а Петя свою - нет.
4. $R_{11}$ &mdash; Вася и Петя решили свои задачи верно

Каждое из этих двух событий несовместно. Более того все четыре вместе они образуют *полную группу событий*.

По условия задачи нам необходимо найти вероятность: $P(R_{01} \cup R_{10})$.
Событий $R_{01}$ и $R_{10}$ несовместны, следовательно $P(R_{01} \cup R_{10}) = P(R_{01}) + P(R_{10})$.

Теперь необходимо найти $P(R_{01})$. Это событие состоит из двух назависимых: Вася решил свою задачу неверно и Петя решил свою задачу верно. Следовательно $P(R_{01}) = 0.1 \cdot (1 - 0.15) = 0.1 \cdot 0.85 = 0.085$.

Аналогично находим $P(R_{10}) = 0.15 \cdot (1 - 0.1) = 0.15 \cdot 0.9 = 0.135$.

Тогда $P(R_{01} \cup R_{10}) = 0.085 + 0.135 = 0.22$.

## Задача 5.
В ящике находится 3 новых теннисных мяча и 7, которыми уже играли Из ящика наугад вынимаются два мяча, которыми играют.
После этого мячи возвращают в ящик. Через некоторое время из ящика берут наугад мяч. 
Найти вероятность того, что он будет новым.

### Решение.
Видимо, автор задачи имеет в виду, что если достали новый мячик и им поиграли, то он перестаёт быть новым.

Можно решать почти как предыдущую задачу. Факт того, что из ящика вынули два мячика можно разбить на три несовместных:

1. $R_{20}$ &mdash; оба вынутых мяча - новые. $P(R_{20}) = \frac{3}{10} \cdot \frac{2}{9} = \frac{1}{15}$
2. $R_{02}$ &mdash; оба вынутых мяча - бывшие в употреблении. 
$P(R_{02}) = \frac{7}{10} \cdot \frac{6}{9} = \frac{7}{15}$
3. $R_{11}$ &mdash; один из вынутых мячей новый, а второй - неоченью.
$P(R_{11}) = 1 - P(R_{02}) - P(R_{20}) = 1 - \frac{1}{15} - \frac{7}{15} = \frac{7}{15}$ так как события образуют полную группу и несовместны. 
Можно проверить себя и посчитать по другому 
$P(R_{11}) = \frac{3}{10} \cdot \frac{7}{9} + \frac{7}{10} \cdot \frac{3}{9} = \frac{7}{30} + \frac{7}{30} = \frac{7}{15}$.
Потому что сделать это можно двумя способами - вытянув сначала старый, а затем новый или наоборот.

Искомая вероятность равна сумме произведений вероятности каждого из событий выше на вероятность того, что вытянутый в третий раз мяч окажется новым.

$P = P(R_{20}) \cdot \frac{1}{10} + P(R_{11}) \cdot \frac{2}{10} + P(R_{02}) \cdot \frac{3}{10} = \frac{1}{15} \cdot \frac{1}{10} + \frac{7}{15} \cdot \frac{2}{10} + \frac{7}{15} \cdot \frac{3}{10} = \frac{1}{150} + \frac{14}{150} + \frac{21}{150} = \frac{36}{150} = \frac{6}{25}$

## Задача 6.
По самолёту производится два выстрела. 
Вероятность попадания при первом выстерел равна 0.5, при втором - 0.6.
При одном попадании самолёт будет сбит с вероятностью 0.7, при двух - с вероятностью 1.
Какова вероятность, что самолёт будет сбит?

### Решение.

Аналогично предыдущим задачам. Разбиваем события на несовместные:

1. $S_{00}$ &mdash; оба выстрела - промах. $P(S_{00}) = (1 - 0.5) \cdot (1 - 0.6) = 0.5 \cdot 0.4 = 0.2$
2. $S_{01}$ &mdash; первый выстрел - промах, а второй попадание. $P(S_{01}) = (1 - 0.5) \cdot 0.6 = 0.5 \cdot 0.6 = 0.3$
3. $S_{10}$ &mdash; первый выстрел - попадание, а второй - промах. $P(S_{10}) = 0.5 \cdot (1-0.6) = 0.5 \cdot 0.4 = 0.2$
4. $S_{11}$ &mdash; оба выстрела - попадание. $P(S_{01}) = 0.5 \cdot 0.6 = 0.3$

У нас есть ещё два события. 
* $F_1$ &mdash; самолёт сбит при одном попадании. $P(F_1) = 0.7$.
* $F_2$ &mdash; самолёт сбит при двух попаданиях. $P(F_2) = 1$.

Вероятность того, что самолёт будет сбит:

$P = P(S_{00}) \cdot 0 + P(S_{01}) \cdot P(F_1) + P(S_{10}) \cdot P(F_1) + P(S_{11}) \cdot P(F_2) = 
0.2 \cdot 0 + 0.3 \cdot 0.7 + 0.2 \cdot 0.7 + 0.3 \cdot 1 = 0.65$

## Задача 7.
Четверь стрелков одета в фуражки, каждый из них попадает в цель в 80% случаев,
остальные одеты в кепки и попадают в цель в 50% случаев. 
Выбранный наугад стрелок выстрелил и попал.
Что вероятнее: он одет в кепку или в фуражку?

### Решение.

Будем считать, что событие $A$ - стрелок одет в фуражку, а событие $B$ - в кепку. Событие $C$ - попадание в мишень.

1. Событие $A \cap C$ заключается в том, что выбранный на угад стрелок носил фуражку и попал в цель. 

$P(A \cap C) = P(A) \cdot P(C|A) = 0.25 \cdot 0.8 = 0.2$.

2. Вероятность того, что стрелок. который попал в цель носит фуражку, по опредлению условной вероятности:

$P(A|C) = \frac{P(A \cap C)}{P(C)}$

3. События $A$ и $B$ образуют полную группу событий. Следовательно событие $C$ можно посчтитать по формуле полной вероятности:

$P(C) = P(C|A) \cdot P(A) + P(C|B) \cdot P(B)$

Запишем $P(A|C)$ по-другому, разложив $P(C)$:

$P(A|C) = \frac{P(A \cap C)}{P(C|A) \cdot P(A) + P(C|B) \cdot P(B)}$

P(C|A) = \frac{P(C \cap A)}{P(A)} = 0.8$

$P(A|C) = \frac{P(A \cap C)}{P(C)} = \frac{P(A \cap C)}{P(C|A) \cdot P(A) + P(C|B) \cdot P(B)} = 
\frac{0.2}{0.8 \cdot 0.25 + 0.5 \cdot 0.75} = \frac{0.2}{0.575}$

Аналогичным образом посчитаем $P(B|C)$ вероятность того, что попавший стрелок носил кепку.

$P(B|C) = \frac{P(B \cap C)}{P(C)} = \frac{P(B \cap C)}{P(C|A) \cdot P(A) + P(C|B) \cdot P(B)} =
\frac{P(B) \cdot P(C|B)}{0.575} = \frac{0.75 \cdot 0.5}{0.575} = \frac{0.375}{0.575}$

И так $P(B|C) > P(A|C)$ следовательно вероятнее, что он был одет.


## Задача 8.
Любая лампа перегорает при скачке напряжения с вероятность 0.2.
В коридоре горит 6 ламп. Напряжение скакнуло. С какой вероятностью перегорят три?

### Решение.

Событие, вероятность которого мы пытаемся найти состоит в том, что три лампы перегорели, а три - не перегорели.

Пусть $p = 0.2$ вероятность, что какая-то лампа перегорит. 
Соответственно $q = 1 - p = 0.8$ вероятность, что какая-то лампа не перегорит после скачка.

Представим себе 6 лампочек, построенных в линию: $1 2 3 4 5 6$

Событие "3 лампочки перегорели" может означать, что лампочки $(1, 2, 3)$ перегорели, а может что перегорели лампочки $(2, 5, 6)$. Вероятность каждого из этих отдельных событий: $P' = p ^ 3 \cdot q ^ 3$ 

Но всего различных комбинаций по три: $C_6^3 = 20$

Искомая вероятность: $P = P' \cdot C_6^3 = p^3 \cdot q^3 \cdot C_6^3 = 0.08192$

## Задача 9.
Точка с координатой $x$ выбирается наудачу на отрезке $[0,1]$ и независимот от неё точка координатой $y$ выбирается на отрезке $[0,2]$.
Найдите вероятность события $A = \{ \max(x,y) < \frac{1}{2} \}$

### Решение.

Переформулируем событие $A$.

$A = \{ \max(x,y) < \frac{1}{2} \} = \{ x < \frac{1}{2} \land y < \frac{1}{2} \} = \{ x < \frac{1}{2} \} \cap \{y < \frac{1}{2} \}$ - так как точку x, y мы выбираем независимо

$P(A) = P(\{ x < \frac{1}{2} \}) \cdot P(\{y < \frac{1}{2} \}) = \frac{\frac{1}{2}}{1} \cdot \frac{\frac{1}{2}}{2} = \frac{1}{2} \cdot \frac{1}{4} = \frac{1}{8}$
